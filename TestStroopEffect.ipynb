# Project Overview:

In a Stroop task, participants are presented with a list of words, with each word displayed in a color of ink.The participant’s task is to say out loud the color of the ink in which the word is printed. 

In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.

Congruent words condition:-
In this condition, Colour word Name is displayed in the same coloured ink as that of word name.
for example RED is written RED ink as shown below.

Incongruent words condition:-
In this condition, Colour word Name is not displayed in the same coloured ink as that of word name.
for example RED is written in GREEN ink as shown below.

### Q2: What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

#### Hypothesis Testing

#### Null Hypothesis:
#### H0: μCongruent − μIncongruent = 0

It states that there are no differences between the response time for congruent words (Congruent) vs incongruent words (Incongruent). The mean difference between paired observations is zero.

where 

μCongruent - the population mean of response time for congruent words. 

μIncongruent - the population mean of response time for incongruent words.

#### Alternate Hypothesis:
#### HA: μCongruent − μIncongruent ≠ 0
It states that the difference between the response time for congruent words vs incongruent words is not zero.   

#### Hypothesis Testing Justification:
The appropriate hypothesis choice will be the one which proves and validates the existence of the Stroop effect ; which suggest that the response time for naming the colour is less for congruent words than that of incongruent words. Hence, there must be a difference in the central tendencies of the population from which the sample is derived. 
Here, the question is - is this difference significant enough to be considered?
Since, we cannot take into consideration the entire population, we need to work with the samples at hand and use the method of hypothesis and statistical testing to infer the results.

#### Proposed  Test– Two tailed dependent sample T-test  

#### Why the two tailed dependent sample T-test?

The selection of T-test is due to the fact that, we are comparing reaction time of a sample from same population in two different conditions (conguent & incongruent word conditions) and hence coming to a conclusion about the population. 

If we get a significant result, we can reject the null hypothesis and accept the alternative hypothesis that there are statistically significant differences between the mean time taken to name ink color between two test conditions.

### Q5: Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

In order to test the hypotheses and perform T test, I have further calculated statistical data such as:

```python
# Calculation of difference between both independant variable and its respective stats data

difference = stroop_data['Congruent'] - stroop_data['Incongruent']
difference_summary = difference.describe()
mean_difference = difference_summary.loc['mean']
std_difference = difference_summary.loc['std']

print("\nMean of the difference is {:.4f} s".format(mean_difference))
print("\nStandard deviation of the difference is {:.4f} s".format(std_difference))
```

#### Degree of Freedom

Degree of freedom is the number of values which are free to vary without violating any given restriction.

Degree of freedom = n - 1

where, n = number of pairs in this case = 24

```python
# Degree of freedom Calculation (dfr)
n = 24
dfr = n-1
print("Degree of Freedom : {}".format(dfr))
```

#### Paired t test :

t statistic to test whether the means are different can be calculated as follows:



where 

x¯D is the mean of difference between all the pairs and 

sD is the sample standard deviation of the difference betweeen all the pairs. 

μ0 is The constant μ0 in our case is 0, since H0:μCongruent − μIncongruent = 0

```python
# T test Calculation
sd = std_difference / np.sqrt(24)
t_stats = mean_difference / sd
print("\nt-statistic for the test is {:.4f}".format(t_stats))
```

If the t-statistic lies outside the critical values of the t-distribution corresponding to our confidence level and degrees of freedom, we reject the null hypothesis.

#### t- critical value calculation

Significance level(α) is the criterion used for rejecting null hypothesis. Statisticians have commonly used either the 0.05 level (5% chance) or the 0.01 level (1% chance).

Confidence level is 1 - significance level.

Alternative hypothesis states HA : μCongruent − μIncongruent ≠ 0. 

Since we hypothesized the possibility of relationship in both directions, we will use a two-tailed test to test our hypothesis. If we are using a significance level of 0.05, two-tailed t-test allocates 0.025 in each tail(shaded area). Both left and right shaded area are 2.5% of the total area under the curve.

A two-tailed test will test both, if μCongruent − μIncongruent is significantly greater than or less than μ0(0). 

The μCongruent−μIncongruent is considered significantly different from μ0(0) if the test statistic is in the top 2.5% or bottom 2.5% of its probability distribution, resulting in a p-value less than 0.05.



```python
# t- critical value calculation
t_critical = (stats.t.ppf(q=0.025, df= dfr), stats.t.ppf(q=0.975, df =dfr))
print("\nt_critical values at alpha = 0.05 for two-tailed T test:\n ({:.4f}, {:.4f}).".format(*t_critical))
```

So, here the t_statistic value (-8.0207) < t_crtical values (-2.0687) at alpha = 0.05

It means that the probability of finding t-statistic as extreme as ours is less than 5% if the null hypothesis was true. This probability is defined by the term p-value.

#### p value calculation:

P value is probability of null hypothesis given that the null hypothesis (H0) is true.

The lower the p-value the greater the confidence with which we can reject the null hypothesis.

```python
# P Value Calculation
pval = stats.t.cdf(t_stats, dfr)*2
print("P value is: {:.4e}".format(pval))
```

The p-value is 4.1030e-08  which means we would expect a 0.000004103 chance of finding a difference as large as (or larger than) the one in our study if the null hypothesis was true. 

Our p-value is (0.000004103) too lower than our significance level α (0.05) 

So we should reject the null hypothesis.

#### Confidence Interval Calculation

Confidence intervals (CI) are a useful statistic to include because they indicate where the true population mean might be. It is common to report 95% confidence intervals.


```python
# Calculation of Confidence interval

ci = stats.norm.interval(0.95, loc = mean_difference, scale = sd)
print("Confidence Intervel : {}".format(ci))
```
